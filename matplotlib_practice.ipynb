```python
#Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot - MATLAB
plt.plot(x_axis, y_axis)
```

```python
# Create the plot - OBJECT ORIENTED INTERFACE "ax.plt()""
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot with label, marker and defined color.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

If you want to show a restircted number of months, you will have to index the X-axis
Use plt.xlim() and apply the index such as plt.xlim(0, 5)

To change the figure size use plt.figure(figsize=(XX, xx)) such as "plt.figure(figsize=(20, 10))"

### Scater plot charts

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.scatter(y_axis, x_axis,c="red",label="Chicago")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
plt.legend("Chicago")
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Enlarge the markers in the chart by iterating through all exisiting data points
#for loop
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#Performs same resizing of data points using
#Python's list comprehension technique under "plt.scatter()"

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

### Create a Scatter Plot Using the Object-Oriented Interface

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Bubble chart
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis,c="blue",s=[i*5 for i in y_axis],label="Boston")
ax.invert_yaxis()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax.legend("Boston")
```

### Create Pie charts

```python
#Labels will equal the x axis 
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Adding the explore pie edge
# Use array-like data points for each value in the pie chart.
# The length of the tuple is equal to the number of wedges in the pie chart
#Uses the decimal point to pop-up July
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

#Increases the size of chart
plt.subplots(figsize=(8, 8))

# To add the percentage, apply the "autopct" and provide the format (i.e.: ".1f%")
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(7, 7))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
